```python
SentimentIntensityAnalyzer - to get the neg/neu/pos scores of the text.
```

```python
import nltk
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm

sia = SentimentIntensityAnalyzer()
```

```python
x=df['review_text'].sample(1)
y=sia.polarity_scores(x.values[0])
print(x.values)
print(y)
```

```python
# create a new column for the polarity scores

polarity_scores = []
for i in range(len(df)):
    polarity_scores.append(sia.polarity_scores(df.iloc[i]['review_text']))
df['polarity_scores'] = polarity_scores
return df
```

```python
df['polarity_scores']
```

```python
# Taking only the compound score
compound=[]
for i in df['polarity_scores']:
    compound.append(i['compound'])
df['compound'] = compound
    
```

```python
df
```

```python
# TEXT REVIEW vs Review overall
df.groupby('beer_style')[['review_overall','compound']].mean().reset_index().sort_values('compound', ascending=False)
```

```python
# REVIEW OVERALL vs text review
df.groupby('beer_style')[['review_overall','compound']].mean().reset_index().sort_values('review_overall', ascending=False)
```
